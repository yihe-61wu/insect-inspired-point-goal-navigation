```python
fig, axes = plt.subplots(6, 5, figsize=(8, 8), dpi=300, sharey='row', sharex=True)

shortest_dist = np.linalg.norm(pos_start_goal[:, -1] - pos_start_goal[:, 0])   # 6.449360748630709 # 

trial_idx = np.arange(n_trial) + 1

for axcol, bias in zip(axes.T, ls_bias):

    for model_name, color in zip(model_names, ls_color):
        time_end = []    
        time_collision = []
        time_reflex = []
        end_dist = []
        straight = []
        val_spl = []
        for force, steer in zip(alldata[bias][model_name]['f'], alldata[bias][model_name]['steer']):
            time_collision.append(np.count_nonzero(force > 100))
            time_reflex.append(np.count_nonzero(steer == 'reflex'))

        for x, y in zip(alldata[bias][model_name]['x'], alldata[bias][model_name]['y']):
            path = np.transpose([x, y])
            end_dist.append(np.linalg.norm(data['pos_goal'][:2] - path[-1]))
            straight.append(np.sum(np.linalg.norm(path[:-1] - path[1:], axis=1)) - shortest_dist)
            time_end.append(len(x))
            
            val_spl.append(spl(path[:500], pos_start_goal[:, 1], shortest_dist))

        axcol[0].scatter(trial_idx, end_dist, c=color, label=model_name, s=1)
        axcol[1].scatter(trial_idx, time_end, c=color, label=model_name, s=1)
        axcol[2].scatter(trial_idx, straight, c=color, label=model_name, s=1)

        axcol[3].scatter(trial_idx, time_collision, c=color, label=model_name, s=1)
        axcol[4].scatter(trial_idx, time_reflex, c=color, label=model_name, s=1)
        axcol[5].scatter(trial_idx, val_spl, c=color, label=model_name, s=1)


for ax in axes[0]:
    ax.axhline(0.559 * 0.3 / 2, color='k', ls='dashed', label='robot radius')

    
for ax, tit in zip(axes.T[0], ('final distance', 'total epoch', 'detour distance', 'collision epoch', 'escape epoch', 'SPL')):
    ax.set_ylabel(tit)
    
# axes[0].legend()
# axes[-1].set_xlabel('trial')

for ax in axes.flatten():
    ax.grid()
```
